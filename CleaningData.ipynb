## X·ª≠ L√Ω D·ªØ Li·ªáu Reaction

```python
# L·ª±a ch·ªçn c√°c d·ªØ li·ªáu c·∫ßn ph√¢n t√≠ch
new_df = raw_df[["post_id", "post_text", "time", "likes", "comments", "shares", "reactions", "reaction_count"]]
new_df
```

```python
# T·∫°o 1 list ch·ª©a c√°c dictionary c·ªßa c√°c lo·∫°i reaction c·ªßa t·ª´ng post v√† l∆∞u v√†o 1 list 
react_list = []
for react_str in new_df.reactions:
    react_dict = ast.literal_eval(react_str)
    react_list.append(react_dict)
# T·∫°o dataframe t·ª´ react_list
df_reactions = pd.DataFrame(react_list)
df_reactions
```

```python
# G·ªôp c√°c c·ªôt l·∫°i v√†o dataframe v√† x√≥a c·ªôt reactions c≈©
new_df_reactions = pd.concat([new_df, df_reactions], axis=1).drop('reactions', axis=1)
new_df_reactions
```

```python
# Thay th·∫ø c√°c gi√° tr·ªã NaN b·∫±ng 0
new_df_reactions = new_df_reactions.fillna(0)
new_df_reactions
```

```python
# T·∫°o 1 list ch·ª©a c√°c ng√†y th√°ng ƒëƒÉng b√†i
date_list = []
for date in new_df_reactions.time:
    date_list.append(date[:10])
# √âp ki·ªÉu thu·ªôc t√≠nh time t·ª´ str th√†nh datetime
i = 0
for time in new_df_reactions.time:
    to_date = datetime.strptime(time, '%Y-%m-%d %H:%M:%S')
    new_df_reactions.iloc[i, 2] = to_date
    i += 1
# T·∫°o dataframe ch·ª©a ng√†y v√† th√°ng ƒëƒÉng b√†i
df_date = pd.DataFrame(date_list, columns = ["date"])

# G·ªôp date v√†o dataframe v√† x√≥a time c≈©
new_df_reactions_date = pd.concat([new_df_reactions, df_date], axis=1).drop('time', axis=1)
new_df_reactions_date
```

```python
# X√≥a c·ªôt likes do ƒë√£ c√≥ c·ªôt like t·ª´ reactions t√°ch ra ·ªü tr√™n
reaction_df = new_df_reactions_date.drop('likes', axis=1)
```

```python
# ƒê·ªïi v·ªã tr√≠ c√°c c·ªôt cho thu·∫≠n
new_column_order = ['post_id','post_text','date','reaction_count','comments','shares','like','love','haha','wow','care','angry','sad']
# S·ª≠ d·ª•ng reindex ƒë·ªÉ ƒë·ªïi v·ªã tr√≠ c√°c c·ªôt
reaction_df = reaction_df.reindex(columns=new_column_order)
```

```python
# th√™m index cho dataframe
reaction_df = reaction_df.reset_index(inplace = False)
reaction_df["index"] = reaction_df["index"] + 1
```

```python
# L∆∞u th√†nh file csv
reaction_df.to_csv("CleanFile/SpaceSpeakersCM_Reaction.csv", index = False)
# L∆∞u th√†nh file xlsx
reaction_df.to_excel("CleanFile/SpaceSpeakersCM_Reaction.xlsx", index=False, engine='openpyxl')
```

## X·ª≠ L√Ω D·ªØ Li·ªáu Comment

```python
import json
import re
import datetime
```

```python
# X·ª≠ l√Ω comment
def get_comments(text):
    pattern = r"\{[^{}]*\}"
    matches = re.findall(pattern, text)
    result = [eval(match) for match in matches]
    return result
```

```python
# T·∫°o m·ªôt dataframe cho comment
new_df = raw_df.dropna(subset=['comments_full'])
comment_list = []
for index, row in new_df.iterrows():
    comments = get_comments(row['comments_full'])  
    for comment in comments:
        try:
            comment_dict = {
                'post_id': row['post_id'],
                'comment_id': comment['comment_id'],
                'comment_text': comment['comment_text'],
                'comment_time': comment['comment_time'],
            }
            comment_list.append(comment_dict)
        except:
            continue

comment_df = pd.DataFrame(comment_list)
comment_df
```

```python
# th√™m index cho dataframe
comment_df = comment_df.reset_index(inplace = False)
comment_df["index"] = comment_df["index"] + 1
comment_df
```
